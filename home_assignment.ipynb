```python

from nilearn import plotting, image, datasets
import nibabel as nib
import os
import glob
import matplotlib.pyplot as plt


mni_template = datasets.load_mni152_template()


#locating the files 

folder = '/Users/heddapilgaard/Downloads/Programming class 2025/Home assignment- Programming'

anat_file = '/Users/heddapilgaard/Downloads/Programming class 2025/Home assignment- Programming /anxiety_association-test_z_FDR_0.01.nii'
funct_file =  '/Users/heddapilgaard/Downloads/Programming class 2025/Home assignment- Programming /anxiety_uniformity-test_z_FDR_0.01.nii'

funct_img = nib.load(funct_file)


#plotting the map:

plotting.plot_stat_map(
    stat_map_img=func_img,
    bg_img=mni_template,     
    title="Brain Scan - Anxiety",
    cmap='hot',
    threshold=0.1,
    cut_coords=(-20, -4, -16),
    display_mode='ortho'
)

plotting.show()
```

Plotting a histogram of the values in the functional data file:

```python

file1 = '/Users/heddapilgaard/Downloads/Programming class 2025/Home assignment- Programming /anxiety_association-test_z_FDR_0.01.nii'
file2 =  '/Users/heddapilgaard/Downloads/Programming class 2025/Home assignment- Programming /anxiety_uniformity-test_z_FDR_0.01.nii'

data1 = nib.load(file1).get_fdata().flatten()
data2 = nib.load(file2).get_fdata().flatten()

data1 = data1[data1 != 0]
data2 = data2[data2 != 0]

plt.figure()
plt.hist(data1, bins=100, alpha=0.5, label="Association", color = "pink")
plt.hist(data2, bins=100, alpha=0.5, label="Uniformity", color = "turquoise")
plt.legend()
plt.title("Histogram of Anxiety Values MRI Scan")
plt.xlabel("Z-value")
plt.ylabel("Voxels")
plt.show()
```

Comments describing the reasoning:

The x-axis (z-values), shows the Z-value a voxel has, while the y-axis (voxels) indicate how many voxels have that given Z-value. For instance we see that there are very many voxels with a Z-value between around 3-6, while the rest of the voxels have lower z-values. This means that the brain regions related to anxiety are specific and localized, rather than being distributed over wide brain regions, as can also be demonstrated in the heatmap above. Brighter heat in the heatmap would correlate to the higher Z-values in the histogram. There is however a crucial difference between the heatmap and the histogram, which is the fact that only the heatmap shows where in the brain the concentration lies (spatial), while the histogram only demonstrates the distribution of values, without indicating where in the brain these values are.
